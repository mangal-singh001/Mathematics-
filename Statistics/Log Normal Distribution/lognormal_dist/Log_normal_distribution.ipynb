```python
import pandas as pd  
import seaborn as sns 
import numpy as np 
```

```python
df = pd.read_csv(
    r"F:\Data Science\Maths\Statistics\Distributions\Log Normal Distribution\lognormal_dist\income.csv",
    index_col = None,
    names = ["income","count"],
    skiprows=1
)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(11.7,8.27)})

g = sns.barplot(x='income', y='count', data=df)
g.set_xticklabels(g.get_xticklabels(),
                  rotation=45, 
                  horizontalalignment='right')

plt.show()
```
